Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import RobustScaler
import joblib
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)
```

Load Climate Profile Dataset

```python
df = pd.read_csv("../datasets/processed/climate_profiles.csv")
print("Dataset Loaded Successfully")
print("Shape:", df.shape)
```

Select Features for Anomaly Detection

```python
anomaly_features = [
    "temperature_celsius",
    "humidity",
    "pressure_mb",
    "precip_mm",
    "cloud",
    "uv_index",
    "visibility_km",
    "air_quality_PM2.5",
    "air_quality_PM10",
    "air_quality_Carbon_Monoxide",
    "air_quality_Ozone",
    "air_quality_Nitrogen_dioxide",
    "air_quality_Sulphur_dioxide",
    "day_length_minutes",
    "temperature_gap",
]
X = df[anomaly_features]
print(X.shape)
```

Scale Features

```python
anomaly_scaler = RobustScaler()
X_scaled = anomaly_scaler.fit_transform(X)
print(X_scaled.shape)
```

Save Anomaly Scaler

```python
joblib.dump(anomaly_scaler, "../models/anomaly_scaler.pkl")
print("Anomaly scaler saved")
```

Hyperparameter tuning

```python
for c in [0.005, 0.01, 0.02, 0.03]:
    model = IsolationForest(contamination=c, random_state=42)

    preds = model.fit_predict(X_scaled)

    print(c, np.sum(preds == -1))
```

Train Isolation Forest

```python
iso_forest = IsolationForest(
    n_estimators=200, contamination=0.01, random_state=42, n_jobs=-1
)
iso_forest.fit(X_scaled)
print("Isolation Forest trained")
```

Generate Predictions

```python
df["anomaly_flag"] = iso_forest.predict(X_scaled)
df["anomaly_flag"].value_counts()
```

Create Readable Labels

```python
df["anomaly_status"] = df["anomaly_flag"].map({1: "Normal", -1: "Anomaly"})
df["anomaly_status"].value_counts()
```

Anomaly Percentage

```python
anomaly_percent = (df["anomaly_flag"] == -1).mean() * 100
print(f"Anomaly Percentage: {anomaly_percent:.2f}%")
```

Visualize

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="anomaly_status", data=df)
plt.title("Anomaly Detection Distribution")
plt.show()
```

Investigate Anomalies

```python
anomalies = df[df["anomaly_flag"] == -1]
print("Anomalies Found:")
print(anomalies.shape)
anomalies[anomaly_features].describe()
```

Compare Climate Profiles vs Anomalies

```python
pd.crosstab(df["climate_profile"], df["anomaly_status"], margins=True)
```

Save Model

```python
joblib.dump(iso_forest, "../models/isolation_forest.pkl")
print("Isolation Forest model saved")
```

Save Dataset

```python
df.to_csv(
    "../datasets/processed/climate_anomalies.csv",
    index=False
)
print(
    "Dataset saved successfully"
)
```
