Load D4 Dataset

```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)
df = pd.read_csv(
    "../datasets/processed/D4_datetime_processed.csv"
)
print("D4 Dataset Loaded")
print(df.shape)
```

Temperature Gap

```python
df["temperature_gap"] = (
    df["feels_like_celsius"]
    - df["temperature_celsius"]
)
df["temperature_gap"].describe()
```

Air Quality Difference

```python
df["pm_difference"] = df["air_quality_PM10"] - df["air_quality_PM2.5"]
df["pm_difference"].describe()
```

Pollution Intensity

```python
df["pollution_intensity"] = (
    df["air_quality_PM2.5"] + df["air_quality_PM10"] + df["air_quality_Carbon_Monoxide"]
)
```

Wind Chill Indicator

```python
df["wind_humidity_interaction"] = df["wind_kph"] * df["humidity"]
```

Rainfall Predictor Interaction

```python
df["humidity_cloud_interaction"] = df["humidity"] * df["cloud"]
```

Heatwave Indicator

```python
df["heatwave_index"] = df["temperature_celsius"] * df["uv_index"]
```

Day Period

```python
def get_day_period(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"


df["day_period"] = df["hour"].apply(get_day_period)
df["day_period"].value_counts()
```

Season Feature

```python
def get_season(month):
    if month in [3, 4, 5]:
        return "Summer"
    elif month in [6, 7, 8, 9]:
        return "Monsoon"
    elif month in [10, 11]:
        return "Post_Monsoon"
    else:
        return "Winter"


df["season"] = df["month"].apply(get_season)
df["season"].value_counts()
```

Verify Engineered Features

```python
engineered_features = [
    "temperature_gap",
    "pm_difference",
    "pollution_intensity",
    "wind_humidity_interaction",
    "humidity_cloud_interaction",
    "heatwave_index",
    "day_period",
    "season",
    "day_length_minutes",
]
print(df[engineered_features].head())
```

Save D5 Dataset

```python
import os

os.makedirs("../datasets/engineered", exist_ok=True)

df.to_csv("../datasets/engineered/D5_feature_engineered.csv", index=False)

print("D5 dataset saved successfully")
```
