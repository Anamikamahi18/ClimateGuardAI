Imports

```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)
df = pd.read_csv(
    "../datasets/processed/D3_feature_removed.csv"
)

print("Dataset Loaded")
print(df.shape)
```

Convert last_updated to Datetime

```python
df["last_updated"] = pd.to_datetime(
    df["last_updated"],
    format="%d-%m-%y %H:%M"
)
print(df["last_updated"].dtype)
```

Extract Date Components

```python
df["year"] = df["last_updated"].dt.year
df["month"] = df["last_updated"].dt.month
df["day"] = df["last_updated"].dt.day
df["hour"] = df["last_updated"].dt.hour
df["weekday"] = df["last_updated"].dt.dayofweek

df[["year", "month", "day", "hour", "weekday"]].head()
```

Datetime Feature Report

```python
print("Unique Years:", df["year"].nunique())
print("Unique Months:", df["month"].nunique())
print("Unique Hours:", df["hour"].nunique())
```

Convert Sunrise/Sunset

```python
df["sunrise"] = pd.to_datetime(df["sunrise"], format="%I:%M %p")

df["sunset"] = pd.to_datetime(df["sunset"], format="%I:%M %p")
```

Create Day Length Feature

```python
df["day_length_minutes"] = (df["sunset"] - df["sunrise"]).dt.total_seconds() / 60
df["day_length_minutes"].describe()
```

Convert Moonrise/Moonset


```python
df["moonrise"] = pd.to_datetime(df["moonrise"], format="%I:%M %p", errors="coerce")

df["moonset"] = pd.to_datetime(df["moonset"], format="%I:%M %p", errors="coerce")
```

Verify Missing Values Created


```python
print("Moonrise Missing:", df["moonrise"].isnull().sum())

print("Moonset Missing:", df["moonset"].isnull().sum())
```

Save D4 Output

```python
import os

os.makedirs("../datasets/processed", exist_ok=True)

df.to_csv("../datasets/processed/D4_datetime_processed.csv", index=False)

print("D4 dataset saved successfully")
```
